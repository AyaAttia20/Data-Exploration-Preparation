### Do universities with a high number of applications are the preferred ones by students? in other words, could the number of applications tell us that a university is one of the most preferred by students?


```python
high_unver_app=data.groupby('Name')['Applicants total'].max().sort_values(ascending=False)
print(high_unver_app[:5])
```

```python
df_applicants=pd.DataFrame(high_unver_app[:5])
df_applicants
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df_applicants['Applicants total'],y=df_applicants.index)
```

#### Insight:
The Top_5 Universities with a high number of applications are [  University of California-Los Angele,University of California-Berkeley,University of California-San Diego,New York University,University of California-Irvine]

#### show the total enrollment in each university

```python
enrol=data.groupby('Name')['Enrolled total'].max().sort_values(ascending=False)[:5]
enrol
```

```python
df_enrol=pd.DataFrame(enrol)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=df_enrol['Enrolled total'],y=df_enrol.index)
```

### Insight:
The universities that have higher number of applications, don't have higher enrolled so universities with a high number of applications are't the preferred ones by students.
